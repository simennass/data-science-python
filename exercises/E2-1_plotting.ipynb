# Plotting

Create a series of plots that demonstrate how the distribution of data differs between the datasets containing trafikkregistreringsstasjon data:

```python
import pandas as pd
```

```python
data1 = pd.read_csv('data/trafikkregistreringspunkt_5_-_uke_42_og_43_2020.csv')
data2 = pd.read_csv('data/trafikkregistreringspunkt_6_-_uke_42_og_43_2020.csv')
```

```python
import seaborn as sns
```

```python
sns.histplot(data=data1, x='vehicle_type', log_scale=[False, True], legend=True)
#sns.histplot(data=data2, x='vehicle_type', log_scale=[False, True], legend=True)
```

```python
sns.histplot(data=data2, x='vehicle_type', log_scale=[False, True])
```

```python
data1['valid'] = data1['valid_classification'] & data1['valid_event'] &\
                data1['valid_length'] & data1['valid_speed']
```

```python
data2['valid'] = data2['valid_classification'] & data2['valid_event'] &\
                data2['valid_length'] & data2['valid_speed']
```

```python
sns.barplot(data=data1, x='lane', y='speed', ci='sd')
```

```python
sns.barplot(data=data2, x='lane', y='speed', ci='sd')
```

```python
sns.scatterplot(data=data1, x='length', y='speed', hue='vehicle_type')
```

```python
sns.scatterplot(data=data2, x='length', y='speed', hue='vehicle_type')
```

```python
sns.catplot(data=data1, kind='bar', x='vehicle_type', y='length', hue='valid')
```

```python
sns.catplot(data=data2, kind='bar', x='vehicle_type', y='length', hue='valid')
```

```python
sns.boxplot(data=data1, x='vehicle_type', y='length', hue='valid')
```

```python
sns.boxplot(data=data2, x='vehicle_type', y='length', hue='valid')
```
